```python
#get the temperature values from year 2006 to 2023 as thats what the hapiness report considers
temperature_columns = [f"F{year}" for year in range(2006, 2023)]

#calculate avergae
climate_df['Average_Temperature_2006_2022'] = climate_df[temperature_columns].mean(axis=1)
climate_df[['Country', 'Average_Temperature_2006_2022']].head()
```

```python
from graphviz import Digraph
```

```python
dag = Digraph()
```

```python
dag.node("Happiness", "Happiness (Ladder score)")
dag.node("Income", "Income (Log GDP per capita)")
dag.node("SocialSupport", "Social Support")
dag.node("LifeExpectancy", "Healthy Life Expectancy")
dag.node("Freedom", "Freedom to Make Life Choices")
dag.node("Generosity", "Generosity")
dag.node("Corruption", "Perceptions of Corruption")
dag.node("Temperature", "Average Temperature")
dag.node("Dystopia", "Dystopia + Residual")
```

```python
dag.edge("Income", "Happiness")
dag.edge("SocialSupport", "Happiness")
dag.edge("Happiness", "LifeExpectancy")
dag.edge("Freedom", "Happiness")
dag.edge("Generosity", "Happiness")
dag.edge("Corruption", "Happiness")
dag.edge("Temperature", "LifeExpectancy")
dag.edge("Dystopia", "Happiness")
```

```python
dag.edge("SocialSupport", "Income")
dag.edge("Income", "LifeExpectancy")
dag.edge("Income", "Freedom")
```

### Total Effect Model
$Happiness_i$ ~ $Normal(\mu_iᵢ, \sigma^2²)$  
$\mu_i = \alpha + \beta_1 * Income_i + \beta_2 * Social_i$

**Priors**  
$\alpha$ ~ $Normal(0, 10)$  
$\beta_1$ ~ $Normal(0, 1)$  
$\beta_2$ ~ $Normal(0, 1)$  
$\sigma$ ~ $Exponential(0, 1)$  

### Direct Effect Model
$Happiness_i$ ~ $Normal(\mu_iᵢ, \sigma^2²)$  
$\mu_i = \alpha + \beta_1 * Income_i + \beta_2 * Social_i + \beta_3 * Freedom_i$

**Priors**  
$\alpha$ ~ $Normal(0, 10)$  
$\beta_1$ ~ $Normal(0, 1)$  
$\beta_2$ ~ $Normal(0, 1)$  
$\beta_2$ ~ $Normal(0, 1)$  
$\sigma$ ~ $Exponential(0, 1)$
